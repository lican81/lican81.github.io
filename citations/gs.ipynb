```python
sorted_publications = sorted(author['publications'], key=lambda x: x['bib'].get('pub_year', -9999999), reverse=True)
```

```python
sorted_publications[0]
```

```python
proc_year = 2023

print(f'### {proc_year}')

# Print journal papers
print('#### Journal Papers')
print('<ol>')

for publication in sorted_publications:
    if publication['bib'].get('pub_year') == proc_year:
        authors = publication['bib'].get('author', [])
        title = publication['bib'].get('title', '')
        journal = publication['bib'].get('journal', '')
        volume = publication['bib'].get('volume', '')
        page = publication['bib'].get('pages', '')
        url = publication.get('pub_url', '')

        # if journal is blank, skip
        if not journal:
            continue

        author_names = authors.replace(' and ', ', ')
        author_names = author_names.replace('Can Li', "<em class='first'>Can Li</em>")
        author_names = author_names.replace('C. Li', "<em class='first'>C. Li</em>")

        formatted_publication = f"<li>{author_names}, <a href='{url}'>{title}</a>, <em class='journal'>{journal}</em>, {volume}, {page} ({publication['bib'].get('pub_year', '')})</li>"
        print(formatted_publication)
print('</ol>')

# Print conference papers
print('#### Conference Papers')
print('<ol>')
for publication in sorted_publications:
    if publication['bib'].get('pub_year') == proc_year:
        authors = publication['bib'].get('author', [])
        title = publication['bib'].get('title', '')
        conference = publication['bib'].get('conference', '')
        volume = publication['bib'].get('volume', '')
        page = publication['bib'].get('pages', '')
        url = publication.get('pub_url', '')
        
        if not conference:
            continue

        author_names = authors.replace(' and ', ', ')
        author_names = author_names.replace('Can Li', "<em class='first'>Can Li</em>")
        author_names = author_names.replace('C. Li', "<em class='first'>C. Li</em>")

        formatted_publication = f"<li>{author_names}, <a href='{url}'>{title}</a>, <em class='journal'>{conference}</em>, {page} ({publication['bib'].get('pub_year', '')})</li>"
        print(formatted_publication)


        # <li><u>M. Jiang, K. Shan</u>, X. Sheng, C. Graves, J.P. Strachan, <em class='first'>C. Li</em> and J. P. Strachan, “An efficient synchronous updating memristor-based Ising solver for combinatorial optimization”, <em class="journal">IEEE International Electron Devices Meeting (IEDM)</em>, San Francisco, CA, USA, Dec. 3-7, 2022. </li>
```
